# Question 7: Create a cross-tabulation between 'AGE' and 'PLAYING ROLE'.


```python
# Cross-tabulation between 'AGE' and 'PLAYING ROLE'
pd.crosstab(df['AGE'], df['PLAYING ROLE'])
```

# Question 8: Sort the DataFrame by 'SOLD PRICE' in descending order.


```python
# Sort by 'SOLD PRICE' in descending order
df.sort_values(by='SOLD PRICE', ascending=False)
```

# Question 9: Identify the player who got the maximum premium on the base price.


```python
# Calculate premium and identify the player with the maximum premium
df['Premium'] = df['SOLD PRICE'] - df['BASE PRICE']
df[df['Premium'] == df['Premium'].max()][['PLAYER NAME', 'Premium']]
```

# Question 10: Find players who received the maximum premium on their base price.


```python
# Find players with the maximum premium
max_premium = df['Premium'].max()
df[df['Premium'] == max_premium][['PLAYER NAME', 'Premium']]
```

# Question 11: Calculate the average 'SOLD PRICE' for each age category.


```python
# Average 'SOLD PRICE' for each age category
df.groupby('AGE')['SOLD PRICE'].mean()
```

# Question 12: Calculate the average 'SOLD PRICE' for different playing roles in each age category.


```python
# Average 'SOLD PRICE' for each playing role in each age category
df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean()
```
